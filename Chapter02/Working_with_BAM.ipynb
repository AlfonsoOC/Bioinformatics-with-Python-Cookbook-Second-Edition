# The recipe

```python
#pip install pysam
from collections import defaultdict

import numpy as np

%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt

import pysam
```

```python
#importo el archivo vam que ya esta en mi computadora en la carpeta del programa
bam = pysam.AlignmentFile('NA18489.chrom20.ILLUMINA.bwa.YRI.exome.20121211.bam', 'rb')
```

```python
#primero inspeccionamos los headers
headers = bam.header
for record_type, records in headers.items():
    print (record_type)
    for i, record in enumerate(records):
        if type(record) == dict:
            print('\t%d' % (i + 1))
            for field, value in record.items():
                print('\t\t%s\t%s' % (field, value))
        else:
            print('\t\t%s' % record)
```

```python
#ahora inspeccionaremos solo un record 
#recuerda que esto esta cifrado en cierta manera pero la informacion puede ser util

#0-based
for rec in bam:
    if rec.cigarstring.find('M') > -1 and rec.cigarstring.find('S') > -1 and not rec.is_unmapped and not rec.mate_is_unmapped:
        break
print(rec.query_name, rec.reference_id, bam.getrname(rec.reference_id), rec.reference_start, rec.reference_end)
print(rec.cigarstring)  # nos da datos del alineamiento 
print(rec.query_alignment_start, rec.query_alignment_end, rec.query_alignment_length)
print(rec.next_reference_id, rec.next_reference_start, rec.template_length)
print(rec.is_paired, rec.is_proper_pair, rec.is_unmapped, rec.mapping_quality)
print(rec.query_qualities)
print(rec.query_alignment_qualities)
print(rec.query_sequence)
```

```python
#Now, let's plot the distribution of the successfully mapped positions in a subset of sequences in the BAM file:

counts = [0] * 76 #conteo para las 76 posiciones 
for n, rec in enumerate(bam.fetch('20', 0, 10000000)): # solo para el cromosoma 20 de 0 al nt 10 000 000
    for i in range(rec.query_alignment_start, rec.query_alignment_end):
        counts[i] += 1
freqs = [x / (n + 1.) for x in counts]
fig, ax = plt.subplots(figsize=(16,9))
ax.plot(range(1, 77), freqs)

#podemos ver qu een este caso la "mapeabilidad" de los reads no es uniforme ya que es mala en los extremos 
#y cae un poco en el medio 
```

```python
#the distribution of Phred scores across the mapped part of the reads.

phreds = defaultdict(list) #usamos un diccionario default
for rec in bam.fetch('20', 0, None): #nuevamente cromosoma 20 y todas las posiciones 
    for i in range(rec.query_alignment_start, rec.query_alignment_end):
        phreds[i].append(rec.query_qualities[i])
```

```python
maxs = [max(phreds[i]) for i in range(76)]
tops = [np.percentile(phreds[i], 95) for i in range(76)]    #para esto utilizamos Numpy que son los percentiles
medians = [np.percentile(phreds[i], 50) for i in range(76)] #para esto utilizamos Numpy que son los percentiles
bottoms = [np.percentile(phreds[i], 5) for i in range(76)]  #para esto utilizamos Numpy que son los percentiles

medians_fig = [x - y for x, y in zip(medians, bottoms)] #para que el grfico de stacks sea bueno se debe restar los bottoms a los medios   
tops_fig = [x - y for x, y in zip(tops, medians)] #y sucesivamente
maxs_fig = [x - y for x, y in zip(maxs, tops)]
```

```python
#perform a stacked plot of the distribution of Phred scores per position.
fig, ax = plt.subplots(figsize=(16,9))
ax.stackplot(range(1, 77), (bottoms, medians_fig, tops_fig, maxs_fig))
ax.plot(range(1, 77), maxs, 'k-')
```
